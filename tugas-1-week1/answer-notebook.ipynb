## 1. Grafik fungsi sinus dengan $f = 5Hz$ dan $\theta = \frac{\pi}{2}$

 
Rumus sinyal sinus secara umum

```math
f(t) = A\sin(ωt-φ)
```
dengan

```math
A=\text{amplitudo } (m), \\
ω=\text{kecepatan sudut } (\frac{rad}{s})=2πf=\frac{2π}{T}, \\
t=\text{waktu } (s) 
```

Jika sebuah sinyal memiliki frekuensi 5 Hz, sinyal tersebut akan melakukan 5 kali putaran dalam satu detik (dalam satu putaran terdapat 1 lembah dan satu puncak) sehingga terdapat 5 lembah dan 5 puncak tiap satu detik. <br>
Jika sebuah sinyal memiliki fase awal sebesar θ=π/2 radiant, grafik sinyal tersebut akan dimulai dari sudut π/2  radiant atau jika dikonversi ke sudut 360°, maka grafik akan dimulai dari sudut 90°. Visualisasi dari grafik tersebut dapat ditampilkan sebagai berikut.<br>


````python
# import library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Pembuatan grafik
# Shortcut
sin = np.sin
pi = np.pi

# deklarasi nilai setiap variabel
A = 1               # Amplitudo
f = 5               # Frekuensi (Hz)
theta = pi/2     # Fase awal

# Setting plot untuk grafik
t = np.linspace(0, 2, 1000)
x_t = A * sin(2*pi*f*t - theta)

plt.figure(figsize=(10,2))      # size dalam inci
plt.plot(t, x_t)
plt.title("Grafik fungsi sinusoidal 5Hz dengan Fase awal pi/2")
plt.xlabel("waktu (s)")
plt.show()
````

## 2. Grafik fungsi sinus dengan $f = 10Hz, 20Hz, 30Hz$ dan $\theta = 0$

````python
# pembuatan grafik
# deklarasi nilai setiap variabel
A = 1
f_1 = 10
f_2 = 20
f_3 = 30
theta = 0

# Setting plot untuk grafik
t = np.linspace(0, 1, 1000)                         # t = 1 detik

# Fungsi yang dipakai
x_t_1 = A * sin(2*pi*f_1*t - theta)
x_t_2 = A * sin(2*pi*f_2*t - theta)
x_t_3 = A * sin(2*pi*f_3*t - theta)

# Konfigurasi subplot
fig, axs = plt.subplots(2, 2, figsize=(20, 7))

# Grafik 10Hz
axs[0][0].plot(t, x_t_1, color="green")
axs[0][0].set_title("f = 10Hz")
axs[0][0].set_xlabel("Waktu (s)")
axs[0][0].set_ylabel("Amplitudo")

# Grafik 20Hz
axs[0][1].plot(t, x_t_2, color="blue")
axs[0][1].set_title("f = 20Hz")
axs[0][1].set_xlabel("Waktu (s)")
axs[0][1].set_ylabel("Amplitudo")

# Grafik 30Hz
axs[1][0].plot(t, x_t_3, color="red")
axs[1][0].set_title("f = 30Hz")
axs[1][0].set_xlabel("Waktu (s)")
axs[1][0].set_ylabel("Amplitudo")


plt.tight_layout()
plt.show()
````

## 3. Analisis persamaan berdasarkan ilustrasi grafik

### a. Grafik kiri


Ilustrasi tersebut berisi grafik dengan fungsi step. Terdapat tiga fase dari grafik tersebut.<br>
1. Fase saat y = 0,
2. Fase saat y = 1, dan
3. Fase saat y = 2

Dengan demikian, fungsi dari grafik tersebut adalah:
```math
f(x)= \left\{
\begin{aligned}
    & 0;                &\quad x  < -2 \text{ atau } x >2 \\
    & 1;                &\quad -2 \leq x < -1 \text{ atau } 1 \leq x < 2\\
    & 2;                &\quad -1 \leq x < 1
\end{aligned}
\right.
```

### b. Grafik kanan


Ilustrasi tersebut berisi grafik dengan fungsi ramp dan fungsi step. Terdapat tiga fase dari grafik tersebut.<br>
1. Fase saat grafik menaik dari y = 0 hingga y = 1,
2. Fase saat y = 1, dan
3. Fase saat grafik menurun dari y = 1 hingga y = 0

Dengan demikian, fungsi dari grafik tersebut adalah:
```math
f(x)= \left\{
\begin{aligned}
    &\frac{1}{4}x;         &\quad -4 \leq x < -1 \\
    &1;                    &\quad -1 \leq x < 1 \\
    &-\frac{1}{4}x;         &\quad 1 \leq x < 4
\end{aligned}
\right.
```

## 2. Bangkitkan dan tampilkan grafik dua buah sinyal sinus dengan fungsi tertentu, lalu **jumlahkan** kedua sinyal tersebut dan tampilkan grafiknya

fungsi:
1. $ S_1(t) = \sin (8\pi f_c t) $
2. $ S_2(t) = \sin (5\pi f_c t + 0.5\pi) $

Karena di dalam soal tidak disebutkan nilai $f_c$, maka penulis mengasumsikan nilai $f_c = 5Hz$

````python
# Pembuatan grafik tiap fungsi
# Shortcut 
sin = np.sin
pi = np.pi

# Deklarasi variabel
f_c = 5                         # satuan Hz
t = np.linspace(0, 2, 1000)     # Range dari 0 - 2 sec.  

# Deklarasi fungsi
s_t1 = sin(8 * pi * f_c * t)
s_t2 = sin(5 * pi * f_c * t + 0.5*pi)
s_sum = s_t1 + s_t2

# Konfigurasi sub plot
fig, axs = plt.subplots(3, 1, figsize=(15,10))
plt.grid(True)

# Grafik s_1(t)
axs[0].set_ylim(-2,2)
axs[0].set_title("Grafik Fungsi S_1(t)")
axs[0].set_xlabel("waktu (s)")
axs[0].set_ylabel("Amplitudo")
axs[0].plot(t, s_t1, color="green")

# Grafik s_1(t)
axs[1].set_ylim(-2,2)
axs[1].set_title("Grafik Fungsi S_2(t)")
axs[1].set_xlabel("waktu (s)")
axs[1].set_ylabel("Amplitudo")
axs[1].plot(t, s_t2, color="blue")

# Grafik penjumlahan kedua fungsi sebelumnya
axs[2].set_ylim(-2,2)
axs[2].set_title("Grafik Fungsi S_sum(t)")
axs[2].set_xlabel("waktu (s)")
axs[2].set_ylabel("Amplitudo")
axs[2].plot(t, s_sum, color="red")

plt.tight_layout()
plt.show()
````
