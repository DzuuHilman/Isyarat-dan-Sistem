## 1. Sistem

**Sistem** dapat didefinisikan sebagai suatu interkoneksi yang menghubungkan sekumpulan komponen dengan terminal atau *port* sehingga beragam informasi, energi, atau materi dapat dimasukkan ke dalamnya dan diberi perlakuan olehnya.

Terdapat dua klasifikasi sistem:<br>
a. Sistem waktu diskrit<br>
b. Sistem waktu kontinu

### a. Sistem waktu kontinu

Sistem waktu ini berkaitan dengan bentuk representasi matematik yang menggambarkan keseluruhan sistem tersebut dalam keseluruhan waktu.

Umumnya, sistem ini menggunakan operasi integral dalam mencari penyelesaian fungsi pembangikitnya. Contohnya sebagai berikut:
1. Sistem arus di dalam rangkaian RLC (resistor, induktor dan kapasitor)
2. Sistem pergerakan mobil berdasarkan gaya yang diberikan 

### b. Sistem waktu diskrit

Sistem waktu diskrit berkaitan dengan bentuk representasi matematik yang menggambarkan sistem dengan pengambilan sampel waktu di beberapa titik tertentu.

Berbeda dengan sistem kontinu, sistem ini menggunakan operasi substitusi nilai ke dalam fungsi sehingga dapat dihasilkan sebuah grafik patah-patah dengan rentang jarak tertentu. Contohnya sebagai berikut:
1. Sistem pembayaran *interest*(bunga) pada bank
2. Sistem prediksi jumlah pengunjung tiap bulan
3. Sistem penghitungan mata kuliah favorit tiap semester

## 2. Karakteristik Sistem


### a. Sistem Kausal dan Non-Kausal



- **Sistem Kausal:** Sistem disebut kausal jika output pada waktu $( t )$ hanya bergantung pada input di waktu sekarang atau waktu sebelumnya, yaitu $( t \geq 0 )$. Artinya, sistem tidak memperhitungkan nilai input di masa depan. Misalnya, jika $( x(t) )$ didefinisikan untuk $( t \geq 0 )$, output sistem bergantung pada nilai $( x )$ dari waktu sekarang dan sebelumnya.

- **Sistem Non-Kausal:** Sistem disebut non-kausal jika output pada waktu $( t )$ bergantung pada nilai input di masa depan (yaitu, $( t < 0 )$). Dengan kata lain, sistem memperhitungkan atau bergantung pada input di masa depan, yang dapat menjadi tidak praktis dalam sistem waktu nyata.



### b. Sistem Linear



- **Addivitas:** Sistem disebut aditif jika output untuk penjumlahan dua input sama dengan penjumlahan output dari masing-masing input secara terpisah. Secara matematis, jika respons sistem terhadap $( x_1(t) )$ adalah $( y_1(t) )$ dan terhadap $( x_2(t) )$ adalah $( y_2(t) )$, maka respons terhadap $( x_1(t) + x_2(t) )$ haruslah $( y_1(t) + y_2(t) )$.

- **Homogenitas (atau Skala):** Sistem disebut homogen jika skala input dengan faktor $( c )$ menghasilkan output yang diskalakan dengan faktor yang sama $( c )$. Jika $( x(t) )$ menghasilkan output $( y(t) )$, maka $( c \cdot x(t) )$ harus menghasilkan $( c \cdot y(t) )$.



### c. Sistem Waktu-Invarian



- **Waktu-Invarian:** Sistem disebut waktu-invarian jika pergeseran pada sinyal input menghasilkan pergeseran yang identik pada sinyal output. Dengan kata lain, jika input $( x(t) )$ digeser oleh $( t_0 )$, output juga akan bergeser oleh $( t_0 )$. Bentuk output tetap sama; hanya posisinya pada sumbu waktu yang berubah. Misalnya, jika $( y(t) = x(t) + 2 )$, maka jika input digeser oleh $( t_0 )$, outputnya menjadi $( y(t) = x(t - t_0) + 2 )$, yang merupakan versi geser dari output asli.


```python
import numpy as np
import matplotlib.pyplot as plt

# Time range
t = np.linspace(0, 10, 100)

# Input signal x(t) - a simple sine wave
x = np.sin(t)

# Time-Invariant System: y(t) = x(t) + 2
y_invariant = x + 2

# Time-Variant System: y(t) = t * x(t)
y_variant = t * x

# Time shift (e.g., shift by 2 units)
t_shifted = t - 2
x_shifted = np.sin(t_shifted)

# Output for shifted input (Time-Invariant System)
y_invariant_shifted = x_shifted + 2

# Output for shifted input (Time-Variant System)
y_variant_shifted = t_shifted * x_shifted

# Plotting the results
plt.figure(figsize=(12, 8))

# Plot original and shifted outputs for the time-invariant system
plt.subplot(2, 2, 1)
plt.plot(t, y_invariant, label="y(t) = x(t) + 2")
# plt.plot(t_shifted, y_invariant_shifted, label="y(t-t0) for shifted x(t)")
plt.title("Time-Invariant System")
plt.xlabel("Time (t)")
plt.ylabel("Output")
plt.legend()

# Plot original and shifted outputs for the time-variant system
plt.subplot(2, 2, 2)
plt.plot(t, y_variant, label="y(t) = t * x(t)")
# plt.plot(t_shifted, y_variant_shifted, label="y(t-t0) for shifted x(t)")
plt.title("Time-Variant System")
plt.xlabel("Time (t)")
plt.ylabel("Output")
plt.legend()

# Plot input signal
plt.subplot(2, 2, 3)
plt.plot(t_shifted, y_invariant_shifted, label="y(t-t0) for shifted x(t)")
plt.title("Time-Invariant System")
plt.xlabel("Time (t)")
plt.ylabel("Output")
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(t_shifted, y_variant_shifted, label="y(t-t0) for shifted x(t)")
plt.title("Time-Invariant System")
plt.xlabel("Time (t)")
plt.ylabel("Output")
plt.legend()

plt.tight_layout()
plt.show()
```

1. Apakah kausalitas?
   Jawaban: **ya**

   output yang dihasilkan (yaitu arus masuk ke dioda dan resistor, $i(t)$) hanya dipengaruhi oleh input (yaitu tegangan sumber daya, $v(t)$) di waktu bersamaan atau waktu sebelumnya. 

   Hal ini disebabkan karena output (arus listrik) tidak akan mengalir sebelum input (tegangan) diterapkan pada rangkaian. Jadi, sistem ini bersifat kausal.
2. Apakah linear?
   Jawaban: **tidak**

   Komponen dioda memiliki dua fungsi, yaitu bias maju (*forward bias)* dan bias terbalik (*reverse bias*). Salah satu dari kedua fungsi ini akan aktif jika dioda diberi tegangan. Jika fungsi bias maju aktif, tegangan yang masuk ke dalam dioda akan disesuaikan. Bila tegangan sumber $V$ lebih besar dari tegangan maju $V_f$ (tegangan minimum agar arus dapat mengalir melalui dioda), arus listrik dapat mengalir, tetapi tegangan tersebut akan mengalami penurunan sebesar $V_f$.

   >Jika $V > V_f$, maka arus mengalir dengan $V_{dioda} = V - V_f$
   
   Dalam kasus di soal ini, kutub positif dari dioda (anoda) terhubung langsung dengan kutub positif sumber tegangan sehingga fungsi yang aktif pada dioda adalah bias maju. Karena diperlukan batas minimum tegangan agar arus listrik dapat mengalir, hasil sistem rangkaian ini bukan merupakan sistem linear, melainkan sistem non-linear.

   Misalkan terdapat dua input yang berbeda $V_1$ dan $V_2$. Sistem dengan kedua input tersebut tidak akan selalu memenuhi kriteria addivitas karena ada kemungkinan salah satu dari kedua input tersebut bernilai kurang dari $V_f$ sehingga output yang dihasilkan bernilai 0. Hal ini bertentangan dengan karakteristik sistem linear.

   > Addivitas: <br> Jika $V_1$ menghasilkan $I_{t1}$ dan $V_2$ menghasilkan $I_{t2}$ <br> maka <br> $V_1 + V_2$ seharusnya menghasilkan $I_{t1}+I_{t2}$ <br> Padahal ada kemungkinan bahwa $V_1$ atau $V_2 < V_f$ sehingga $I_{t1}$ atau $I_{t2}= 0$ yang melanggar sifat linearitas


3. Apakah *time-invariant*? 
   Jawaban: ya

   Rangkaian tersebut tidak mengalami perubahan selama waktu berlangsung rangkaian bersifat statis (Di asumsi awal, rangkaian tersebut adalah rangkaian DC). Hasil ouput saat ini akan selalu sama dengan hasil output di masa depan.

### a. $y[n] = 2y[n-1];$ dengan $y[-1] = 1$

### b. $y[n] = 0.5y[n-1] + y[n-2];$ dengan $y[-2] = 1, y[-1] = 0$

```python
# Buat fungsi rekursi untuk mencari hasil dari y_n
def y_nb(n):
    if n == -2:
        return 1
    elif n == -1:
        return 0
    else:
        return 0.5 * y_nb(n - 1) + y_nb(n - 2)
    
# Hitung nilai y[n] untuk 0 <= n <= 10
results_b = [y_nb(n) for n in range(11)]

# Sajikan dalam tabel data
df_result_b = pd.DataFrame({
    'n':range(11),
    'y[n]':results_b
})
df_result_b.set_index('n', inplace=True)
df_result_b
```

### c. $y[n] = 0.1y[n-1] + 0.5y[n-2] + (0.5)^n;$ dengan $y[-1]=y[-2]=0$

```python
# Buat fungsi rekursi untuk mencari hasil dari y_n
def y_nc(n):
    if n == -2 or n == -1:
        return 0
    else:
        return 0.1 * y_nc(n - 1) + 0.5 * y_nc(n - 2) + (0.5) ** n

results_c = [y_nc(n) for n in range(11)]

# Sajikan dalam tabel data
df_result_c = pd.DataFrame({
    'n':range(11),
    'y[n]':results_c
})
df_result_c.set_index('n', inplace=True)
df_result_c
```

```python
# Diode in AC circuit

import numpy as np
import matplotlib.pyplot as plt

# Parameters
V_m = 10        # Peak voltage of the AC signal
V_f = 0.7       # Forward voltage of the diode (e.g., 0.7V for a silicon diode)
t = np.linspace(0, 4*np.pi, 1000)  # Time array
omega = 1       # Angular frequency

# Input signal
x_t = V_m * np.sin(omega * t)

# Output signal after passing through the diode (with non-linearity)
y_t = np.where(x_t > V_f, x_t - V_f, 0)

# Plotting the input and output signals
plt.figure(figsize=(10, 6))

plt.plot(t, x_t, label="Input Signal $x(t) = V_m \sin(\omega t)$", color='blue')
plt.plot(t, y_t, label="Output Signal $y(t)$ after Diode", color='red')

plt.title("Non-Linear Behavior of Diode in an AC Circuit")
plt.xlabel("Time (t)")
plt.ylabel("Voltage (V)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Diode in DC circut

import numpy as np
import matplotlib.pyplot as plt

# Parameters
V_DC = 5        # DC supply voltage
V_f = 0.7       # Forward voltage of the diode
R = 1           # Resistor value (ohms)

# DC input signal (constant value)
t = np.array([0, 1])  # Time array (only for demonstration)
v_in = np.array([V_DC, V_DC])  # Constant DC voltage

# Output voltage calculations
v_out_forward = np.maximum(v_in - V_f, 0)  # Forward bias output
v_out_reverse = np.zeros_like(v_in)        # Reverse bias output

# Plotting the DC input and output signals
plt.figure(figsize=(10, 6))

# Forward Bias
plt.plot(t, v_in, label="DC Input Signal $V_{DC}$", color='blue', linestyle='--')
plt.plot(t, v_out_forward, label="Output Signal with Diode (Forward Bias)", color='red')

# Reverse Bias (Assuming a reverse bias scenario)
plt.plot(t, v_out_reverse, label="Output Signal with Diode (Reverse Bias)", color='green', linestyle=':')

plt.title("Behavior of Diode in a DC Circuit")
plt.xlabel("Time (t)")
plt.ylabel("Voltage (V)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
t = np.linspace(-1, 2, 1000)  # Time array
x_t = np.heaviside(t, 1)      # Input: Unit step function

# Impulse response h(t) = 0.5 for 0 <= t <= 1
h_t = np.where((t >= 0) & (t <= 1), 0.5, 0)

# Convolution of x(t) and h(t) to get y(t)
y_t = np.convolve(x_t, h_t, mode='full')[:len(t)]

# Time array adjustment for convolution
t_y = np.linspace(t[0], t[-1], len(y_t))

# Plotting the input, impulse response, and output signals
plt.figure(figsize=(12, 8))

# Plot input signal
plt.subplot(3, 1, 1)
plt.plot(t, x_t, label="Input Signal $x(t)$")
plt.title("Input Signal")
plt.xlabel("Time (t)")
plt.ylabel("Amplitude")
plt.grid(True)

# Plot impulse response
plt.subplot(3, 1, 2)
plt.plot(t, h_t, label="Impulse Response $h(t)$", color='red')
plt.title("Impulse Response")
plt.xlabel("Time (t)")
plt.ylabel("Amplitude")
plt.grid(True)

# Plot output signal
plt.subplot(3, 1, 3)
plt.plot(t_y, y_t, label="Output Signal $y(t)$", color='green')
plt.title("Output Signal")
plt.xlabel("Time (t)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.legend()
plt.show()
```
